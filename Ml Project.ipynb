# Feature Engineering

```python
df["Renovated"] = df["Renovated"].map({"Yes":1, "No":0})
```

```python
df.head(2)
```

```python
df = pd.get_dummies(
    df,
    columns=["Neighborhood" ,"Condition"],
    drop_first=True
)
```

```python
df.head(2)
```

```python
df.dtypes
```

```python
x = df.drop("Price" , axis = 1)
y = df["Price"]
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split( x , y , test_size=0.2 , random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

# Baseline model

```python
from sklearn.metrics import mean_squared_error
```

```python
baseline_pre = np.full(len(y_test), y_train.mean())
baseline_rmse = np.sqrt(mean_squared_error(y_test , baseline_pre))
baseline_rmse
```

# Linear Regresssion and Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
lr.fit(x_train , y_train)
```

```python
y_pred = lr.predict(x_test_scaled)

rmse_lr = np.sqrt(mean_squared_error(y_test , y_pred))
r2_lr = r2_score(y_test, y_pred)

rmse_lr , r2_lr
```

# Ridge method and model

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge(alpha=1.0)
ridge.fit(x_train_scaled , y_train)

y_pred_ridge = ridge.predict(x_test_scaled)

rmse_ridge = np.sqrt(mean_squared_error(y_test,y_pred_ridge))
r2_rige = r2_score(y_test , y_pred_ridge)

rmse_ridge , r2_rige
```
